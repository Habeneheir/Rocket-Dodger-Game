```python
import turtle
import random
import time

# Set up the screen
screen = turtle.Screen()
screen.bgcolor("black")
screen.title("Rocket Dodger Game")
screen.setup(width=600, height=600)  # Set the screen dimensions

# Create a rocket turtle
rocket = turtle.Turtle()
rocket.shape("triangle")
rocket.color("red")
rocket.penup()
rocket.goto(0, -200)
# Create the rocket body (rectangle)
rocket_body = turtle.Turtle()
rocket_body.shape("square")
rocket_body.color("blue")
rocket_body.shapesize(stretch_wid=1, stretch_len=2.5)  # Adjust the size
rocket_body.penup()
rocket_body.goto(rocket.xcor() - 10, rocket.ycor()-30)

# Create the rocket legs 
leg1 = turtle.Turtle()
leg1.shape("triangle")
leg1.color("gray")
leg1.shapesize(stretch_wid=0.5, stretch_len=1)  # Adjust the size
leg1.penup()
leg1.goto(rocket.xcor() - 5, rocket.ycor() - 20)

leg2 = turtle.Turtle()
leg2.shape("triangle")
leg2.color("gray")
leg2.shapesize(stretch_wid=0.5, stretch_len=1)  # Adjust the size
leg2.penup()
leg2.goto(rocket.xcor() + 5, rocket.ycor() - 20)
# Create obstacles
obstacles = []

def create_obstacle():
    obstacle = turtle.Turtle()
    obstacle.shape("square")
    obstacle.color("orange")
    obstacle.shapesize(stretch_wid=1, stretch_len=3)
    obstacle.penup()
    x = random.randint(-300, 300)
    y = random.randint(250, 300)
    obstacle.goto(x, y)
    obstacles.append(obstacle)
    
# Function to move the rocket up
def move_rocket_up():
    y = rocket.ycor() + 20  # Adjust the speed of the rocket
    if -200 < y < 200:
        rocket.sety(y)
        rocket_body.sety(y - 30)
        leg1.sety(y - 60)
        leg2.sety(y - 60)

# Function to move the rocket left
def move_rocket_left():
    x = rocket.xcor() - 20  # Adjust the speed of the rocket
    if -200 < x < 200:  # Check if the new position is within the horizontal border limits
        rocket.setx(x)
        rocket_body.setx(x)
        leg1.setx(x - 10)
        leg2.setx(x + 10)

# Function to move the rocket right
def move_rocket_right():
    x = rocket.xcor() + 20  # Adjust the speed of the rocket
    if -200 < x < 200:  # Check if the new position is within the horizontal border limits
        rocket.setx(x)
        rocket_body.setx(x)
        leg1.setx(x - 10)
        leg2.setx(x + 10)
    
# Function to move the rocket down
def move_rocket_down():
    y = rocket.ycor() - 20  # Adjust the speed of the rocket
    if -200 < y < 200:  # Check if the new position is within the vertical border limits
        rocket.sety(y)
        rocket_body.sety(y - 30)
        leg1.sety(y - 60)
        leg2.sety(y - 60)
# Set the heading to point upwards
rocket.setheading(90)
rocket_body.setheading(90)
leg1.setheading(-90)
leg2.setheading(-90)

# Initialize obstacle speed
obstacle_speed = 2
last_speed_increase_time = time.time()

# Function to move obstacles down
def move_obstacles():
    global obstacle_speed, last_speed_increase_time
    try:
        if screen._root is None:
            # Screen has been closed, stop the function
            return

        for obstacle in obstacles:
            y = obstacle.ycor()
            y -= obstacle_speed  # Adjust the speed of the obstacles
            obstacle.sety(y)
            if y < -200:  # If obstacle is off the screen, reset it
                x = random.randint(-200, 200)
                y = random.randint(300, 300)
                obstacle.goto(x, y)
                
# Increase obstacle speed at regular intervals
        current_time = time.time()
        if current_time - last_speed_increase_time >= 5:  # Increase speed every 5 seconds
            obstacle_speed += 1
            last_speed_increase_time = current_time
          
        # Check for collisions
        check_collisions()

        screen.update()
        screen.ontimer(move_obstacles, 10)  # Repeat the function every 10 milliseconds

    except turtle.Terminator:
        # Handle Terminator exception (e.g., screen closed)
        pass

# Function to check for collisions
def check_collisions():
    for obstacle in obstacles:
        if (
            rocket.xcor() - 20 < obstacle.xcor() < rocket.xcor() + 20
            and rocket.ycor() - 20 < obstacle.ycor() < rocket.ycor() + 20
        ):
            game_over()


# Function to end the game
def game_over():
    global text_turtle  # Make 'text_turtle' a global variable
    # Hide rocket and obstacles
    rocket.hideturtle()
    for obstacle in obstacles:
        obstacle.hideturtle()
# Set up the screen
    #screen = turtle.Screen()
    screen.bgcolor("black")
    screen.title("Game Over Screen")

# Create a turtle for writing
    text_turtle = turtle.Turtle()
    text_turtle.color("white")
    text_turtle.hideturtle()
    text_turtle.penup()

# Write the first text
    text_turtle.write("GAME OVER", align="center", font=("Arial", 21, "normal"))

# Set the position for the second text
    text_turtle.sety(text_turtle.ycor() - 30)  # Adjust the y-coordinate as needed

# Clear previous text and write the second text
    text_turtle.write("Press 'r' to Restart", align="center", font=("Arial", 8, "normal"))

# Keep the window open
    turtle.done()


    screen.onkeypress(reset_game, "r")
    
# Function to reset the game
def reset_game():
    
    global obstacles
    screen = turtle.Screen()
    screen.title("Rocket Dodger Game")
    rocket.goto(0, -200)
    rocket.showturtle()
    text_turtle.clear()
    for obstacle in obstacles:
        obstacle.hideturtle()
    obstacles = []
    for _ in range(10): 
        create_obstacle()
    move_obstacles()
# Keyboard binding
screen.listen()
screen.onkey(move_rocket_up, "Up")
screen.onkey(move_rocket_left, "Left")
screen.onkey(move_rocket_right, "Right")
screen.onkey(move_rocket_down, "Down")
screen.onkey(reset_game, "r")
# Main game loop
for _ in range(10):  # Create 5 initial obstacles
    create_obstacle()
move_obstacles()  # Start moving obstacles

# Close the window when clicking on the 'X'
screen.exitonclick()
```
